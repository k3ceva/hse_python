```python
# Задание 1
# Напишите функцию, которая принимает на вход строку и проверяет, является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2–3 цифры).
# Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

# Если номер валиден, то функция должна возвращать отдельно номер и регион.

# Примеры работы программы:
# car_id = 'А222BС96'
# Результат: Номер А222BС валиден. Регион: 96.

# car_id = 'АБ22ВВ193'
# Результат: Номер не валиден.

def car_id_validator(car_id):
    result = re.match("([ABEKMHOPCTYXАВЕКМНОРСТУХ]{1}[0-9]{3}[ABEKMHOPCTYXАВЕКМНОРСТУХ]{2})([0-9]{2,3})", car_id)
    if result is None:
        print("Номер не валиден.")
    else:
        print(f"Номер {result.group(1)} валиден. Регион: {result.group(2)}.")


car_id_validator('А222BС96')
car_id_validator('АБ22ВВ193')
```
