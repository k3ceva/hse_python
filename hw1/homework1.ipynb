```python
# Задание 1
# Переведите содержимое файла purchase_log.txt в словарь purchases вида:
# {'1840e0b9d4': 'Продукты', ...}
# Пример работы программы при выводе первых двух элементов словаря purchases:
# 1840e0b9d4 ‘Продукты‘
# 4e4f90fcfb  ‘Электроника‘

import json
```

```python
purchases = dict()

with open('purchase_log.txt', encoding='utf-8') as purchase_log:
    purchase_log.readline() # skip header
    s = purchase_log.readline()
    while s != '':
        js = json.loads(s)
        purchases[js['user_id']] = js['category']
        s = purchase_log.readline()
```

```python
# Вывод результата для задания 1
i = 0
print('Первые элементы словаря purchases:')
for k in purchases.keys():
    print(f'{k} ‘{purchases[k]}‘')
    i += 1
    if i == 5:
        break
```

```python
# Задание 2 (необязательное)
# Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки.
# Если покупка была, сам файл visit_log.csv изменять не надо.
# Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
# Учтите условия для данных:
# ●	содержимое purchase_log.txt помещается в оперативную память компьютера;
# ●	содержимое visit_log.csv — нет, используйте только построчную обработку этого файла.

# Примеры работы программы при выводе первых трёх строк файла funnel.csv:
# user_id,source,category
# 1840e0b9d4,other,Продукты
# 4e4f90fcfb,context,Электроника

import csv
```

```python
with open('visit_log (2).csv', encoding='utf-8') as visit_log:
    visit_log_csv = csv.reader(visit_log)
    header = visit_log_csv.__next__()
    user_id_index = header.index('user_id')
    source_index = header.index('source')
    with open('funnel.csv', 'w', encoding='utf-8') as funnel:
        funnel_csv = csv.writer(funnel, lineterminator='\n')
        funnel_csv.writerow(['user_id', 'source', 'category'])
        for row in visit_log_csv:
            user_id = row[user_id_index]
            source = row[source_index]
            category = purchases.get(user_id)
            if category is not None:
                funnel_csv.writerow([user_id, source, category])
```

```python
# Вывод результата для задания 2
print('Первые строки файла funnel.csv:')
with open('funnel.csv', encoding='utf-8') as funnel:
    for line in funnel.readlines(200):
        print(line, end='')
```
